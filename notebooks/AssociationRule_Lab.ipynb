```
from google.colab import drive
drive.mount('/content/drive')

import pandas as pd
import numpy as np
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori, association_rules

# --- config / parameter ---
CSV_PATH       = "/content/drive/My Drive/Data science/Market_Basket_Optimisation.csv"
MIN_SUPPORT    = 0.003     # min support
MIN_CONFIDENCE = 0.15      # min confidence
MIN_LIFT       = 1.00      # >1 = เกิดร่วมกันมากกว่าบังเอิญ
```

```
# 1) โหลดข้อมูล
basketData = pd.read_csv(CSV_PATH)
print("shape =", basketData.shape)
print(basketData.head(3))

# 2) แปลงเป็น list of transactions
dataset = []
for i in range(basketData.shape[0]):
    temp = []
    for j in range(basketData.shape[1]):
        if not pd.isna(basketData.iat[i, j]):
            s = str(basketData.iat[i, j]).strip()
            if s and s.lower() != "nan":
                temp.append(s)
    dataset.append(temp)

print("ตัวอย่าง transaction แรก:", dataset[0][:10])
```

```
# 3) One-hot encoding
te = TransactionEncoder()
te_ary = te.fit(dataset).transform(dataset)
df_onehot = pd.DataFrame(te_ary, columns=te.columns_)

# 4) หา frequent itemsets
frequent_itemsets = apriori(df_onehot, min_support=MIN_SUPPORT, use_colnames=True)
print("frequent_itemsets found:", len(frequent_itemsets))
frequent_itemsets.head()
```

```
# 5) สร้างกฎ
rules = association_rules(frequent_itemsets,
                          metric="confidence",
                          min_threshold=MIN_CONFIDENCE)
rules = rules[rules["lift"] >= MIN_LIFT].copy()

rules["ante_len"] = rules["antecedents"].apply(len)
rules["cons_len"] = rules["consequents"].apply(len)

rules_sorted = rules.sort_values(
    by=["lift", "confidence", "support"],
    ascending=False
)

def pick_rule(df, a_len, b_len):
    sub = df[(df["ante_len"]==a_len) & (df["cons_len"]==b_len)]
    if sub.empty:
        return None
    return sub.iloc[0]

rule1 = pick_rule(rules_sorted, 1, 1)
rule2 = pick_rule(rules_sorted, 2, 1)
rule3 = pick_rule(rules_sorted, 2, 2)

def as_list(fs):
    return ", ".join(list(fs)) if hasattr(fs, "__iter__") else str(fs)

selected = [r for r in [rule1, rule2, rule3] if r is not None]

print("=== Selected rules ===")
if not selected:
    print("ไม่พบกฎที่ผ่านเกณฑ์")
else:
    for i, r in enumerate(selected, 1):
        A = as_list(r["antecedents"])
        B = as_list(r["consequents"])
        print(f"\nRule {i}: {A} -> {B}")
        print(f"  Support    = {r['support']:.4f} ({r['support']:.2%})")
        print(f"  Confidence = {r['confidence']:.4f} ({r['confidence']:.2%})")
        print(f"  Lift       = {r['lift']:.4f}")
        print("-"*60)

if rule3 is None:
    print("\nหมายเหตุ: ไม่พบกฎ (2-itemset)->(2-itemset) ที่ผ่านเกณฑ์ในชุดข้อมูลนี้")
```
